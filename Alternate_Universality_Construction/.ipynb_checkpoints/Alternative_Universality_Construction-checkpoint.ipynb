# Implementation of Alternative Universality Construction 
## (See Problem 4.3 from Nielsen and Chuang for Math Background)

## 1. Import Packages

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import math
import numpy as np
import scipy.linalg as la
from scipy.linalg import logm, expm
from numpy.linalg import matrix_power
from numpy import pi
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info.operators import Operator
```

## 2. Set up Pauli Operators

```python
j = complex(1j)

#Pauli Operators
I = np.array([[1,0],[0,1]])
X = np.array([[0,1],[1,0]])
Y = np.array([[0,0-j],[0+j,0]])
Z = np.array([[1,0],[0,-1]])
```

## 3. Define Functions 

### 3-1. Define H = iln(U)

```python
def get_H(U):
    H = j*logm(U)
    #H = np.around(H,decimals=8)
    return H
```

### 3-2. Get the Set of Pauli Operators (g) according to the number of Qubits (n)

```python
# tensor list of the Pauli operators (Pauli Set)
#n is the number of qubits

def get_PS(n):
    PS = [I,X,Y,Z]
    
    i=1
    while i<=n-1:
        PS_tem = []
        for Op in PS:
            PS_tem.append(np.kron(Op,I))
            PS_tem.append(np.kron(Op,X))
            PS_tem.append(np.kron(Op,Y))
            PS_tem.append(np.kron(Op,Z))
        PS = PS_tem
        i=i+1
    
    return PS
```

```python
#string version of get_PS for the convenience of circuit construction
def get_tensor_list(n):
    TL = ["I","X","Y","Z"]
    
    i=1
    while i<=n-1:
        TL_tem = []
        for Op in TL:
            TL_tem.append(Op+str("I"))
            TL_tem.append(Op+str("X"))
            TL_tem.append(Op+str("Y"))
            TL_tem.append(Op+str("Z"))
        TL = TL_tem
        i=i+1
    
    return TL
```

### 3-3. Find hg –– H as a Linear Combination of the Pauli Operators (g)

```python
#find hg by solving system of linear equations
def get_hg(U):
    H = get_H(U)
    n = math.log(H.shape[0],2)
    PS = get_PS(n)
    
    coeffs = list([])
    RHS = list([])
    
    #coeffs*x = RHS
    for matrix in range(4**int(n)):
        for r in range(2**int(n)):
            for c in range(2**int(n)): 
                coeffs.append(PS[matrix][r][c])
        
    #reshape then transpose
    coeffs = np.reshape(coeffs,(4**int(n),4**int(n))).T

    
    #RHS(right hand side of the equations) is the matrix elements of H
    for r in range(2**int(n)):
            for c in range(2**int(n)): 
                RHS.append(H[r][c])
    

    
    hg = np.linalg.solve(coeffs,RHS)
    

    return hg
```

### 3-4. Get the Approximated Form of U (see Problem 4.3-(5)) 

```python
def get_U(U,k):
    if Operator(U).is_unitary():
        delta = 1/k
        hg = get_hg(U)
        H = get_H(U)
        n = math.log(H.shape[0],2)
        PS = get_PS(n)

        u=expm(-j*hg[0]*PS[0]*delta)
        for i in range(1,4**int(n)):
            u = u.dot(expm(-j*hg[i]*PS[i]*delta))
        U_approx = matrix_power(u,k)
    else:
        print('The Matrix is not Unitary')
        
    return U_approx
```

### 3-5. Construct Circuit according to hg 

```python
def construct_circuit_U(U,k):
    repeat = 0
    hg = get_hg(U)
    H = get_H(U)
    n = math.log(H.shape[0],2)
    TL = get_tensor_list(n)
    
    #create a list for the positions of TL of non-zero hg terms
    num=0
    list_num = []
    for term in hg:
        if term!=0:
            list_num.append(num)
        num = num+1
    
    qc = QuantumCircuit(n)
    while repeat<k:
        for i in list_num:
            tensor_product = TL[i][::-1]
            
            non_I_list = []
            #pauli is the position of the pauli operator
            for pauli in range(int(n)):
                if tensor_product[pauli]!="I":
                    non_I_list.append(pauli)
            
            #X,Y basis change (see Nielsen and Chuang chap 4.7)
            for pauli in non_I_list:
                if tensor_product[pauli]=="X":
                    qc.h(pauli)
                    #qc.u3(pi/2,0,pi,pauli)
                if tensor_product[pauli]=="Y":
                    qc.rx(pi/2,pauli)
                    #qc.u3(pi/2,-pi/2,pi/2,pauli)
            
            #Ladder
            for c in range(len(non_I_list)-1):
                if c+1<=len(non_I_list):
                    qc.cx(non_I_list[c],non_I_list[c+1])         
                else:
                    break
                       
            #Exp
            if len(non_I_list)>0:
                qc.rz(hg[i].real*2,non_I_list[-1])
                #qc.u3(0,0,hg[i].real*2,non_I_list[-1])
        
            #Ladder
            for c in range(len(non_I_list)-1,0,-1):
                if c-1>=0:
                    qc.cx(non_I_list[c-1],non_I_list[c])         
                else:
                    break           
            

            #,Y,Z basis change (see Nielsen and Chuang chap 4.7)
            for pauli in non_I_list:
                if tensor_product[pauli]=="X":
                    qc.h(pauli)
                    #qc.u3(pi/2,0,pi,pauli)
                if tensor_product[pauli]=="Y":
                    qc.rx(-pi/2,pauli)
                    #qc.u3(-pi/2,-pi/2,pi/2,pauli)
                      
        repeat = repeat + 1
    
    #qc=transpile(qc,basis_gates=['u3','cx'],optimization_level=3)

    return qc
```
